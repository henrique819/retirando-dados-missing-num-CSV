```python
#Importando o arquivo CSV para um dataframe
import pandas as pd

#fazendo a exibição de seu conteúdo
tabela = pd.read_csv("vendas_linha_petshop.csv", sep=";" , encoding="ISO-8859-1")
display(tabela)
```

```python
#Identificando as colunas aonde estão os dados missing e a quantidade de dados missing
print(tabela.isnull().sum())
```

```python
#Exibindo apenas os dados missing
tabela[tabela.isnull().any(axis=1)]
```

```python
#Convertendo a tipagem dos tipos de dados de cada atributo
tabela.dtypes
```

```python
#Substituindo os valores missing por 0
tabela['quantidade'].fillna('0', inplace = True)
```

```python
#Confirmando que não há mais dados missing
print(tabela.isnull().sum())
```

```python
#Visualizando a nova tabela sem os dados missing (substituido por 0)
display(tabela)
```

```python
#Exportando o dataframe sem missing para um CSV
tabela.to_csv('nomissing_vendas_linha_petshop.csv')
```
